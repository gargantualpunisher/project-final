# Outliers detection and removal 

```python
import numpy as np
```

```python
mean = np.mean(data1)
std = np.std(data1)
threshold = 3
z_scores = (data1 - mean) / std
outliers = data1[np.abs(z_scores) > threshold]
df = data1[~data1.isin(outliers)].dropna()
```

```python
df = df.reset_index()
df1 = df.drop(['index'], axis=1)
```

```python
df1
```

# encoding

```python
from sklearn.preprocessing import LabelEncoder
label = LabelEncoder()
df1['STATE/UT'] = label.fit_transform(df1['STATE/UT'])
df1['DISTRICT'] = label.fit_transform(df1['DISTRICT'])
```

```python
df1
```

# Normalisation of data

```python
from sklearn.preprocessing import StandardScaler, RobustScaler
from sklearn.preprocessing import PowerTransformer
```

```python
ss = RobustScaler()
scaled= pd.DataFrame(ss.fit_transform(df1))
```

```python
scaled
```

# after Normalisation spread of data

```python

for column in scaled.columns:
    if scaled[column].dtype != 'object':
        plt.figure(figsize=(8, 6))
        plt.scatter(scaled[column], range(len(scaled)), marker='.')
        plt.title(f"Scatter Plot - {column}")
        plt.xlabel(column)
        plt.ylabel('Index')
        plt.show()
```

# Elbow Method

```python
from sklearn.cluster import KMeans
```

```python
elbow = pd.DataFrame()

for i in range(1,15):
    kmean = KMeans(n_clusters=i, random_state=333)
    model = kmean.fit(scaled)
    elbow.at[i,'k'] = i
    elbow.at[i, 'inertia'] = round(model.inertia_)/10000000
print(elbow)
    
```

```python
plt.figure(figsize=(15,8))
plt.title('Elboe plot', size=20)
plt.plot(elbow['k'], elbow['inertia'])
plt.xticks(elbow['k'])
plt.xlabel('k', size=15)
plt.ylabel('inertia', size=15)
```

# Model Building

```python
df = pd.DataFrame()
kmeans = KMeans(n_clusters = 4,  random_state = 42)

y_kmeans = kmeans.fit_predict(scaled)

df['y_kmeans']=y_kmeans

df['y_kmeans'].value_counts()
```

# accuracy score

```python
from sklearn.metrics import silhouette_score


silhouette_avg = silhouette_score(scaled, y_kmeans)
print("Silhouette Coefficient:", silhouette_avg)
```
